```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda1 = 0.15
lambda2 = 0.75
N10 = 1
N20 = 0
t = np.linspace(0, 16, 300)

# Solutions
N1 = N10 * np.exp(-lambda1 * t)
# N2 = (lambda1 * N10 / (lambda2 - lambda1)) * (np.exp(-lambda1 * t) - np.exp(-lambda2 * t))
N2 = N20 * np.exp(-lambda2 * t) + (lambda1 * N10 / (lambda2 - lambda1)) * (np.exp(-lambda1 * t) - np.exp(-lambda2 * t))

# Plot
plt.plot(t, N1, label="$N_1$(t)", color="blueviolet")
plt.plot(t, N2, label="$N_2$(t)", color="limegreen")
plt.xlabel("time, t")
plt.ylabel("number of particles, N")
plt.title("Decay of $N_1$ and $N_2$")
plt.legend()
plt.grid(True)
plt.show()
```
